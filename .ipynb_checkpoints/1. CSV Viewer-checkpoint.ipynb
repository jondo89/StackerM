```python
import pandas as pd
import plotly.graph_objects as go

FileName = '18-04-2023_06-59-07-992.csv'

# Load the CSV file into a Pandas dataframe
df = pd.read_csv('C:/Users/jdavi/Desktop/Lumpy April/d/'+FileName, delimiter=';', encoding="utf-8-sig")

# Create the figure
fig = go.Figure()

# Add a trace
fig.add_trace(go.Scatter(
    x=df.index, 
    y=df['Channel 2.2'],
    mode='lines',
    name='Channel 2.2'
))
# Add a trace
fig.add_trace(go.Scatter(
    x=df.index, 
    y=df['Channel 2.1'],
    mode='lines',
    name='Channel 2.1'
))
# Add a trace
fig.add_trace(go.Scatter(
    x=df.index, 
    y=df['Channel 2.3'],
    mode='lines',
    name='Channel 2.3'
))
# Add a trace
fig.add_trace(go.Scatter(
    x=df.index, 
    y=df['Channel 2.4'],
    mode='lines',
    name='Channel 2.4'
))
# Set plot title and axis labels
fig.update_layout(
    title=FileName,
    xaxis_title='Time',
    yaxis_title='Value',
)

# Show the plot
fig.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
FileName = '04-06-2022_03-41-15-525.csv'
# Load the CSV file into a Pandas dataframe
df = pd.read_csv('./csv/'+FileName, delimiter=';', encoding="utf-8-sig")

# View the first few rows of the dataframe

# Plot the time series data
plt.plot(df.index, df['Channel 2.2'])

# Set plot title and axis labels
plt.title(FileName)
plt.xlabel('Time')
plt.ylabel('Value')

# Show the plot
plt.show()
```
